### Import Library

### Load Data

Data diambil dari https://data.jakarta.go.id/dataset/indeks-standar-pencemaran-udara-ispu-tahun-2020

```python
data_train = pd.read_csv('data_train.csv')
data_train
```

Keterangan Fitur dan Label

1. tanggal : Tanggal pengukuran kualitas udara
2. stasiun : Lokasi pengukuran di stasiun
3. pm10 : Partikulat salah satu parameter yang diukur
4. so2 : Sulfida (dalam bentuk SO2) salah satu parameter yang diukur
5. co : Carbon Monoksida salah satu parameter yand diukur
6. o3 : Ozon salah satu parameter yang diukur
7. no2 : NItrogen dioksida salah satu parameter yang diukur
8. max : Nilai ukur paling tinggi dari seluruh parameter yang diukur dalam waktu yang sama
9. critical : Parameter yang hasil pengukurannya paling tinggi
10. categori : Kategori hasil perhitungan indeks standar pencemaran udara

### Pre-Processing Data

```python
#Cek info data dan tipe variabelnya
data_train.info()
```

```python
#Drop kolom tanggal karena fitur tanggal dianggap kurang relevan
data_train.drop(columns='tanggal', inplace=True)
```

```python
#Cek missing values
data_train.isnull().sum()
```

Terdapat 8 missing values pada fitur "critical"

```python
#Menyimpan index yang mengandung missing value pada fitur critical
index_nan_critical = np.where(data_train['critical'].isnull())[0]
index_nan_critical
```

```python
#Melihat data yang mengandung missing value pada fitur critical berdasarkan index pada index_nan_critical
for i in index_nan_critical:
    print(data_train.iloc[i])
```

Dapat dilihat bahwa nilai Null pada fitur 'critical' ternyata data untuk fitur lain juga kosong

```python
# Dimensi data awal dataframe
print("Ukuran awal data: {baris} baris, {kolom} kolom.".format(baris=data_train.shape[0], kolom=data_train.shape[1]))

# Drop baris jika ada satu saja data yang missing dengan function dropna() 
data_train = data_train.dropna()

#Dimensi data tanpa missing values
print("Ukuran data tanpa missing value: {baris} baris, {kolom} kolom.".format(baris=data_train.shape[0], kolom=data_train.shape[1]))
```

```python
#Fungsi untuk mengecek nilai unik pada tiap kolom
def check_unique(data):
    for col in data.columns:
        print(col, data[col].unique())
```

```python
#Melihat nilai unik pada tiap kolom fitur
check_unique(data_train)
```

Dapat dilihat masih ada beberapa data yang memiliki value '---' pada fitur **'pm10', 'so2', 'co', '03', dan 'no2'**. Kami akan mengganti data tersebut dengan median untuk menghindari jumlah data yang sedikit nantinya

```python
#Mengganti semua value '---' dengan Nan
data_train = data_train.replace('---', np.nan)
```

```python
#Mengecek kembali data yang memiliki missing value
data_train.isnull().sum()
```

```python
#Mengganti semua nilai "nan" dengan nilai median pada tiap fiturnya
data_train = data_train.fillna(data_train.median())
```

```python
#Mengecek kembali unique value dan missing value
data_train.isnull().sum()
```

```python
#Mengecek kembali nilai unik pada tiap kolom fitur
check_unique(data_train)
```

Dapat dilihat data sudah bersih dari **'missing value'**, selanjutnya akan dilakukan eksplorasi untuk mengetahui **'outliers'**

```python
#Cek info data dan tipe variabelnya
data_train.info()
```

Kolom **pm10, so2, co, o3, no2** masih bertipe object. Kami mengubah tipe data fitur tersebut menjadi integer, agar bisa dilakukan eksplorasi

```python
#Mengubah tipe data fitur pm10, so2, co2, o3, no2 menjadi int
data_train['pm10'] = data_train['pm10'].astype(int)
data_train['so2'] = data_train['so2'].astype(int)
data_train['co'] = data_train['co'].astype(int)
data_train['o3'] = data_train['o3'].astype(int)
data_train['no2'] = data_train['no2'].astype(int)
```

```python
#Cek info data dan tipe variabel setelah mengubah tipe data pada kolom fitur pm10, so2, co2, o3, no2
data_train.info()
```

```python
# Membersihkan kolom stasiun
data_train['stasiun']= data_train['stasiun'].apply(lambda x: x.split("(")[0])
data_train.head()
```

```python
# Statistik data
data_train.describe()
```

```python
# Melihat distribusi data
data_train.hist(figsize=(10,10))
plt.show()
```

Dari grafik diatas dapat dikatakan distribusi data tidak beraturan

```python
#Deteksi outliers menggunakan boxplot
f, axes = plt.subplots(6,figsize=(10,20))
sns.boxplot(data_train['pm10'], data=data_train, ax=axes[0])
sns.boxplot(data_train['so2'], data=data_train, ax=axes[1])
sns.boxplot(data_train['co'], data=data_train, ax=axes[2])
sns.boxplot(data_train['o3'], data=data_train, ax=axes[3])
sns.boxplot(data_train['no2'], data=data_train, ax=axes[4])
sns.boxplot(data_train['max'], data=data_train, ax=axes[5])
```

Berdasarkan Boxplot diatas dapat terlihat beberapa kolom fitur memiliki nilai data yang outliers, akan diterapkan metode IQR untuk menghindari outliers

```python
Q1 = data_train.quantile(0.25)
Q3 = data_train.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
data_train.shape
```

```python
data_train = data_train[~((data_train < (Q1 - 1.5 * IQR)) |(data_train > (Q3 + 1.5 * IQR))).any(axis=1)]
data_train.shape
```

Setelah dilakukan imputasi outlier dengan nilai median dapat dilihat tidak ada lagi outlier pada kolom

```python
#Cek kembali distribusi data
data_train.hist(figsize=(10,10))
plt.show()
```

Dapat dilihat Setelah dilakukan imputasi outlier, distribusi data jadi normal

```python
#Melakukan reset index agar index menjadi terurut
data_train = data_train.reset_index(drop=True)
```

```python
#Mengubah fitur stasiun dan critical dari data categorical menjadi numerical menggunakan library label encoder
from sklearn.preprocessing import LabelEncoder 
labelencoder = LabelEncoder()

#Mengubah fitur stasiun menjadi numerical
data_train['stasiun'] = labelencoder.fit_transform(data_train['stasiun'])

#Mengubah fitur critical menjadi numerical
data_train['critical'] = labelencoder.fit_transform(data_train['critical'])
```

```python
#Melihat 5 data pertama untuk mengetahui hasil label encoder pada kolom fitur stasiun dan critical
data_train.head()
```

```python
#Melihat jumlah masing-masing class label pada class target
data_train['categori'].value_counts()
```

```python
#Mengubah class target menjadi numerical
categori = {'TIDAK SEHAT':0, 'BAIK': 1, 'SEDANG': 2, }
data_train = data_train.replace({'categori':categori})
```

```python
#Menyimpan values data pada variabel baru (berbentuk numpy array)
data_values = data_train.values
```

```python
#Split data menjadi input dan output
X, y = data_values[:, :-1], data_values[:, -1]
```

Dikarenakan datanya imbalanced, maka akan dilakukan over sampling terhadap data class label yang minoritas

```python
#Melihat jumlah masing-masing class label pada class target yang sudah diubah menjadi numerical
data_train['categori'].value_counts()
```

```python
#Import library SMOTE
from imblearn.over_sampling import SMOTE
```

```python
#Import library Counter
from collections import Counter
```

```python
#Melakukan oversampling pada class label yang menjadi minoritas
oversample = SMOTE()
X, y = oversample.fit_resample(X, y)
```

```python
#Memastikan panjang data input dan output sama

#Mengecek panjang data input
len(X) == len(y)
```

```python
#Mengubah X kembali menjadi dataframe 'data_balanced'
clean_data_train = pd.DataFrame(X).reset_index(drop=True)
```

```python
#Melihat 5 baris data pertama pada "data_balanced"
clean_data_train.head()
```

Dikarenakan dataframe "data_balanced" belum memiliki header maka akan ditambakan header(nama kolom) pada dataframe "data_balanced"

```python
#Mengisi header pada dataframe "data_balanced"
headers = ['stasiun', 'pm10', 'so2', 'co', 'o3','no2','max','critical']
clean_data_train.columns=headers
```

```python
#Menambahkan nilai y pada datafrane "data_balanced" dengan nama kolom "categori"
clean_data_train['categori'] = pd.DataFrame(y).reset_index(drop=True)
```

```python
#Melihat kembali 5 data pertama pada dataframe "data_balanced"
clean_data_train.head()
```

```python
# Dimensi data awal dataframe
print("Ukuran awal data: {baris} baris, {kolom} kolom.".format(baris=data_train.shape[0], kolom=data_train.shape[1]))

#Dimensi data tanpa missing values
print("Ukuran data setelah oversampling: {baris} baris, {kolom} kolom.".format(baris=clean_data_train.shape[0], kolom=clean_data_train.shape[1]))
```
